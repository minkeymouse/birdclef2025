```python
import numpy as np
import pandas as pd

meta = pd.read_csv("../data/birdclef/DATABASE/train_metadata.csv")
bad = []

for path in meta["label_path"].unique():
    arr = np.load(path)
    if np.isnan(arr).any():
        bad.append(path)

print("These label files contain NaN:", bad)
```

```python
import numpy as np, pandas as pd
import os
meta = pd.read_csv("../data/birdclef/DATABASE/train_metadata.csv")
bad = [p for p in meta["label_path"].unique() if not os.path.exists(p)]
print("Missing label .npy files:", bad[:100])
```
